```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_color_codes=True
```

```python
car = pd.read_csv('data\car.csv')
# car.head(10)
```

```python
## Data set

# Gets data type
data_types = car.dtypes
# Gets columns
all_columns = car.columns
# Analytical summary of data set
data_sumary = car.describe(include='all')
data_sumary
```

```python
# Plotting

car.hist(figsize=(8,10))
```

```python
# Relationship between categorical and continues variables
sns.boxplot(x='Vehicle Size', y='Engine HP', data=car)
```

```python
#sns.pairplot(car)
```

```python
# Drop columns and always specify axis
car = car.drop(['Engine Fuel Type','Market Category','Vehicle Style','Popularity','Number of Doors','Vehicle Size'],axis=1)
car.head(10)
```

```python
# Renaming Columns
car = car.rename(columns={'Engine HP':'HP','Engine Cylinders':'Cylinders'})
car.head(10)
```

```python
# Total number of rows and columns
car.shape
```

```python
# Duplicate rows
duplicate_rows = car[car.duplicated()]
duplicate_rows.shape

# Count the rows before removing data
car.count()

# Drop the duplicates
car.drop_duplicates()
car.head(10)
```

```python
# Drop the duplicates
car.drop_duplicates()
car.head(10)

# Check count again
car.count()
```

```python
# Find all null values
# If u do noot put the sum, it will return a mask of true or false
car.isnull().sum()
```

```python
# Drop the missing values
car = car.dropna()
car.count()
```

```python
# Check null values again
car.isnull().sum()
```

```python
# Plotting a histogram for the number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
```

```python
# Find the relationship between variables
plt.figure(figsize=(20,10))
c = car.corr()
sns.heatmap(c,cmap='BrBG',annot=True)
c
# The closer to one, the more positive the correlation between the variable
# The heatmap shows how cose the variables are correlated. The darker the colour the more correlated they are
```
